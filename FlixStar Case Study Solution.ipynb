```python
#This code will drop the columns mentioned in the list from the dataset. 

movies = movies.drop(['color', 
                      'director_facebook_likes', 
                      'actor_3_facebook_likes', 
                      'actor_1_facebook_likes', 
                      'cast_total_facebook_likes', 
                      'actor_2_facebook_likes', 
                      'duration', 
                      'facenumber_in_poster', 
                      'content_rating', 
                      'country', 
                      'movie_imdb_link', 
                      'aspect_ratio',
                      'plot_keywords',
                      'actor_2_name',
                      'actor_3_name'], 
                       axis = 1)
movies
```

-  ### Subtask 2.3:

#### write here what the following codeblock  is doing ( 1 mark )

```python
#The first line will discard all the rows with null values in the column gross and 
#keep only the rows with non-null values in the original dataset

#The second line will discard all the rows with null values in the column budget and 
#keep only the rows with non-null values in the original dataset

movies = movies[~np.isnan(movies['gross'])]
movies = movies[~np.isnan(movies['budget'])]
movies
```

-  ### Subtask 2.4: 

#### write here what the following codeblock  is doing ( 1 mark )

```python
#This code will calculate the total null values in each columns
#Then it will only keep the columns in which the total null values are less than 5
#Basically the code is discarding all the columns which have more than 5 null values
movies = movies[movies.isnull().sum(axis=1) <= 5]
movies
```

-  ### Subtask 2.5: 

#### write here what the following codeblock is doing ( 1 mark )

```python
#This is code replacing all the rows in the column language with null value, with the value English

movies['language'].describe()
movies.loc[pd.isnull(movies['language']), ['language']] = 'English'
movies
```

- ### Subtask 3.4: Display the names of  the top 5 directors based on IMDb ratings with a minimum of 5 movies (Top Directors Playlist)  (4 marks)

```python
# Write your code here
#Algorithm
#1. Getting the average IMDb score of all directors
#2. Get the count of movies for each director
#3. From this, filter the directors who have directed at least 5 movies
#4. From these filtered directors, get the top 5 directors based on IMDb rating
```

```python
#Step 1
top_movies_director = pd.DataFrame(movies.groupby('director_name')['imdb_score'].mean())

#Step 2
top_movies_director['Movies_Count'] = movies['director_name'].value_counts().sort_index()

#Step 3
top_movies_director = top_movies_director[top_movies_director['Movies_Count'] >= 5]

#Step 4
top_movies_director = top_movies_director.sort_values(by='imdb_score',ascending=False)

#Final playlist
top5_movies_director_playlist = top_movies_director.head(5)
```
